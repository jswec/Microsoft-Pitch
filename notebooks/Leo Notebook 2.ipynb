# JOIN IMDB CELLS FOR COLLABORATOR INFO

```python
collaborators_raw = movie_titles.merge(principals_prod,
                                       left_on='movie_id',
                                       right_on = 'movie_id',
                                       how ='inner')
collaborators_raw.head()
```

```python
collaborators_raw = collaborators_raw.merge(persons,
                                           left_on= 'person_id',
                                           right_on= 'person_id',
                                           how = 'inner')
collaborators_raw.head()
```

```python
collaborators_raw = collaborators_raw[['movie_id', 'primary_title', 'start_year', 'genres', 'person_id', 'category', 'job', 'primary_name', 'primary_profession']]
collaborators_raw['start_year'].value_counts()
```

```python
collaborators_raw = collaborators_raw[(collaborators_raw['start_year'] <= 2019) & (collaborators_raw['start_year'] > 2015)]
pd.DataFrame(collaborators_raw)
collaborators_raw['start_year'].value_counts()
```

```python
collaborators_raw['category'].value_counts()
```

```python
collaborators_raw = collaborators_raw[(collaborators_raw['category'] == 'actor') | 
                                     (collaborators_raw['category'] == 'actress') |
                                     (collaborators_raw['category'] == 'director') |
                                     (collaborators_raw['category'] == 'producer') |
                                     (collaborators_raw['category'] == 'writer')]
collaborators_raw['category'].value_counts()
```
